To get data for this project I explored carefuly few real estate sites specialised on apartments in Paris. I was looking for hidden API to use it for webscraping without lxml or BeautifulSoup or CSS selectors or XPath. Just nice, simple JSON.

The site www.bienici.com (a French site for finding real estate for buying / renting apartments in France) uses frontend frameworks which render dynamic content by loading a JSON file from the backend to populate the user-facing site.

So my task was to find the URL to access that dynamic conten and to write a nice code using querystring, payload and headers which the site itself uses.
